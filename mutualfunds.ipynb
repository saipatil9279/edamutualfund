```python
#Importing the neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Loading the dataset
df = pd.read_csv('/content/Mutual_Funds.csv')
```

```python
#View the data
df.head
```

```python
# Calculate the mode of SIP Min. Inv.(Rs.)
sip_min_inv_mode = df['SIP Min. Inv.(Rs.)'].mode()[0]

# Print the mode
print("Mode of SIP Min. Inv.(Rs.):", sip_min_inv_mode)
```

```python
#Filling the missing values with the mode using fillna() function
df['SIP Min. Inv.(Rs.)'].fillna(sip_min_inv_mode, inplace=True)

# Save the updated dataframe to a new CSV file
df.to_csv('Mutual_Funds_updated.csv', index=False)
```

```python
#Viewing the newly saved file
df.head
```

```python
#Using pd.to_numeric function to convert values to numeric
# Convert CAGR% 6 Months column to numeric values
df['CAGR% 6 Months'] = pd.to_numeric(df['CAGR% 6 Months'], errors='coerce')

# Replace NaN values with the mean of the column
cagr_6m_mean = df['CAGR% 6 Months'].mean()
df['CAGR% 6 Months'].fillna(cagr_6m_mean, inplace=True)

# Convert CAGR% 1 Year column to numeric values
df['CAGR% 1 Year'] = pd.to_numeric(df['CAGR% 1 Year'], errors='coerce')

# Replace NaN values with the mean of the column
cagr_1y_mean = df['CAGR% 1 Year'].mean()
df['CAGR% 1 Year'].fillna(cagr_1y_mean, inplace=True)

# Convert CAGR% 3 Year column to numeric values
df['CAGR% 3 Year'] = pd.to_numeric(df['CAGR% 3 Year'], errors='coerce')

# Replace NaN values with the mean of the column
cagr_3y_mean = df['CAGR% 3 Year'].mean()
df['CAGR% 3 Year'].fillna(cagr_3y_mean, inplace=True)
```

```python
# Calculate the mean, median, and mode of each column
cagr_6m_mean = df['CAGR% 6 Months'].mean()
cagr_6m_median = df['CAGR% 6 Months'].median()
cagr_6m_mode = df['CAGR% 6 Months'].mode()[0]

cagr_1y_mean = df['CAGR% 1 Year'].mean()
cagr_1y_median = df['CAGR% 1 Year'].median()
cagr_1y_mode = df['CAGR% 1 Year'].mode()[0]

cagr_3y_mean = df['CAGR% 3 Year'].mean()
cagr_3y_median = df['CAGR% 3 Year'].median()
cagr_3y_mode = df['CAGR% 3 Year'].mode()[0]

# Print the results
print("CAGR% 6 Months: Mean = {:.2f}, Median = {:.2f}, Mode = {:.2f}".format(cagr_6m_mean, cagr_6m_median, cagr_6m_mode))
print("CAGR% 1 Year: Mean = {:.2f}, Median = {:.2f}, Mode = {:.2f}".format(cagr_1y_mean, cagr_1y_median, cagr_1y_mode))
print("CAGR% 3 Year: Mean = {:.2f}, Median = {:.2f}, Mode = {:.2f}".format(cagr_3y_mean, cagr_3y_median, cagr_3y_mode))
```

```python
# Count the number of null values in each column
null_counts = df.isnull().sum()

# Print the results
print("Null counts:")
print(null_counts)
```

```python
#Data Cleaning Exp Ratio Column
# Convert Exp. Ratio(%) column to numeric values
df['Exp. Ratio(%)'] = pd.to_numeric(df['Exp. Ratio(%)'], errors='coerce')

# Replace NaN values with the mean of the column
exp_ratio_mean = df['Exp. Ratio(%)'].mean()
df['Exp. Ratio(%)'].fillna(exp_ratio_mean, inplace=True)
```

```python
#Data Cleaning Benchmarks Column
# Remove leading and trailing whitespaces in 'Benchmark' column
df['Benchmark'] = df['Benchmark'].str.strip()

# Replace 'N/A' values with NaN
df['Benchmark'].replace('N/A', np.nan, inplace=True)

# Drop rows with missing values in the 'Benchmark' column
df = df.dropna(subset=['Benchmark'])

# Create a new dataset with the filtered DataFrame
new_dataset = df.copy()

# Optional: Reset the index if needed
new_dataset = new_dataset.reset_index(drop=True)

# Save the new dataset as a CSV file
new_dataset.to_csv('New_Mutual_Funds2.csv', index=False)

print("New dataset saved as 'New_Mutual_Funds2.csv'.")
```

```python
df = pd.read_csv('/content/New_Mutual_Funds2.csv')
# Make sure 'Net_Asset_Value(Rs.)' column is numeric (convert if necessary)
df['Net_Asset_Value(Rs.)'] = pd.to_numeric(df['Net_Asset_Value(Rs.)'], errors='coerce')

# Find the row with the minimum 'Net_Asset_Value(Rs.)'
min_row = df[df['Net_Asset_Value(Rs.)'] == df['Net_Asset_Value(Rs.)'].min()]

# Find the row with the maximum 'Net_Asset_Value(Rs.)'
max_row = df[df['Net_Asset_Value(Rs.)'] == df['Net_Asset_Value(Rs.)'].max()]

# Access the 'Scheme' column for both the minimum and maximum schemes
min_scheme = min_row['Scheme'].values[0]
max_scheme = max_row['Scheme'].values[0]

# Print the results
print(f"Scheme with minimum Net_Asset_Value(Rs.): {min_scheme}")
print(f"Scheme with maximum Net_Asset_Value(Rs.): {max_scheme}")
```

```python
# Find the scheme with maximum CAGR% 6 Months
max_cagr_6_months = df['CAGR% 6 Months'].max()
df_max_cagr_6_months = df[df['CAGR% 6 Months'] == max_cagr_6_months]

# Find the scheme with minimum Exp. Ratio(%)
min_exp_ratio = df_max_cagr_6_months['Exp. Ratio(%)'].min()
scheme_combined = df_max_cagr_6_months[df_max_cagr_6_months['Exp. Ratio(%)'] == min_exp_ratio]['Scheme'].values[0]

print("Scheme with maximum CAGR% 6 Months and minimum Exp. Ratio(%):", scheme_combined)
```

```python
# Find the scheme with maximum CAGR% 1 Year
max_cagr_1_year = df['CAGR% 1 Year'].max()
df_max_cagr_1_year = df[df['CAGR% 1 Year'] == max_cagr_1_year]

# Find the scheme with minimum Exp. Ratio(%)
min_exp_ratio = df_max_cagr_1_year['Exp. Ratio(%)'].min()
scheme_combined = df_max_cagr_1_year[df_max_cagr_1_year['Exp. Ratio(%)'] == min_exp_ratio]['Scheme'].values[0]

print("Scheme with maximum CAGR% 1 Year and minimum Exp. Ratio(%):", scheme_combined)
```

```python
# Find the scheme with maximum CAGR% 3 Year
max_cagr_3_year = df['CAGR% 3 Year'].max()
df_max_cagr_3_year = df[df['CAGR% 3 Year'] == max_cagr_3_year]

# Find the scheme with minimum Exp. Ratio(%)
min_exp_ratio = df_max_cagr_3_year['Exp. Ratio(%)'].min()
scheme_combined = df_max_cagr_3_year[df_max_cagr_3_year['Exp. Ratio(%)'] == min_exp_ratio]['Scheme'].values[0]

print("Scheme with maximum CAGR% 3 Year and minimum Exp. Ratio(%):", scheme_combined)
```
